```python
import numpy as np
import matplotlib.pyplot as plot

np.random.seed(42)
data = np.random.normal(loc=0, scale=1, size=300)

mean = np.mean(data)
median = np.median(data)

plot.figure(figsize=(10, 7))

plot.hist(data, bins=30, alpha=0.7, color='orange', edgecolor='black')

# Add mean and median lines
plot.axvline(mean, color='blue', linestyle='dashed', linewidth=2, label=f'Mean: {mean:.2f}')
plot.axvline(median, color='green', linestyle='dashed', linewidth=2, label=f'Median: {median:.2f}')

# Setting the annotation positions
mean_x_annotation_pos = mean + 0.1
mean_y_annotation_pos = max(np.histogram(data, bins=30)[0]) - 3

median_x_annotation_pos = median + 0.1
median_y_annotation_pos = max(np.histogram(data, bins=30)[0]) - 5

# Annotate mean and median
plot.text(mean_x_annotation_pos, mean_y_annotation_pos, f'Mean: {mean:.2f}', color='blue', fontsize=14)
plot.text(median_x_annotation_pos, median_y_annotation_pos, f'Median: {median:.2f}', color='green', fontsize=14)

plot.title('Histogram of Random Data', fontsize=16)
plot.xlabel('Numeric Data', fontsize=14)
plot.ylabel('Frequency', fontsize=14)

plot.xticks(fontsize=12)
plot.yticks(fontsize=12)

plot.grid(True)

# Add legend
plt.legend()

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plot

parquet_file_url = 'https://star.herts.ac.uk/~kuhn/DHV/exercises_problem2.parquet'

# Reads the data into pandas dataframe
dataframe = pd.read_parquet(parquet_file_url)

# Displays few of the rows of the file
dataframe.head()

# Provides the datatypes present, statistics of the data
print(dataframe.describe())

dataframe.hist(figsize=(14, 10),  bins=30, edgecolor='black')
plot.suptitle('Histogram of Numerical Columns')
plot.show()
```

```python
import seaborn as sns
plot.figure(figsize=(12, 8))
sns.heatmap(dataframe.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plot.title('Correlation Heatmap')
plot.show()

# Pairplot for pairwise relationships
sns.pairplot(dataframe)
plot.suptitle('Pairplot of Data', y=1.02)
plot.show()
```

```python
# Boxplot for each numerical col
plot.figure(figsize=(14, 10))
dataframe.boxplot()
plot.title('Boxplot of Numerical Columns')
plot.xticks(rotation=45)
plot.show()
```
